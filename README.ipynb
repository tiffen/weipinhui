 特征提取
 ---
 1. 用户行为特性：<br>
 用户在距离预测开始时间前3,8,21,34,89天的用户点击次数，购买次数，点击购买率，（最后一次点击、购买距离预测日期的天数：暂时没有）
 2. 商品行为特征：<br>
 用户在距离预测开始时间前3,8,21,34,89天的商品的点击次数，购买次数，点击购买率。
 3. 用户—商品对行为特性：<br>
 用户在距离预测开始时间前3,8,21,34,89天的用户-商品对的点击次数，购买次数，点击购买率。
 4. 品牌的行为特性： <br>
 用户在距离预测开始时间前3,8,21,34,89天的品牌的点击次数，购买次数，点击购买率。品牌对时间不是很敏感，时间区域间距可以大一些。
 5. 用户 -品牌对的行为特性：<br>
 用户在距离预测开始时间前3,8,21,34,89天的用户-品牌对的点击次数，购买次数，点击购买率。
 
 
 数据预处理和采样
 ---

3. 构建的数据集庞大达千万级别，正负样本比例严重失衡，为1:90。为了提高训练速度，又不降低训练准确度，对负样本做五次有放回的下采样，改善正负样本比例为1:10，训练五个模型，用五个模型的预测进行投票产生最终得预测结果。保证在减少了数据量的同时，又防止负样本信息的丢失。但是这样每一个模型都没有获取到所有的训练信息，但是预测的时候却要预测所有的，这样性能会下降，实践证明也是。所以还是通过负样本欠采样80%，正样本过采样500%，改善正负样本比例为1:14左右。

模型训练
---
线下训练数据：对历史数据的倒数第二周的用户-商品对，取倒数第二周之前的所有用户、商品、用户-商品的特征，用倒数第二周内是否购买作为标签。 <br>
线下预测的时候：对历史数据的倒数第一周的用户-商品对，取'01-10'到倒数第一周之前的所有用户、商品、用户-商品的特征，用倒数第一周内是否购买作为标签。 <br><br>

线上训练数据： 对最后一周用户-商品对，取'01-10'到'03-24'的所有用户、商品、用户-商品的特征，用最后一周用户是否购买作为标签。 <br>
线上预测：对预测周的用户-商品对，取'01-18'到'03-31'所有的用户、商品、用户-商品的特征，预测购买的概率，从17号开始是为了与训练时候的数据了保持一致 <br><br>
